```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import iqr

# Load data
fb_data = pd.read_csv("fb_stock_prices_2018.csv")
earthquake_data = pd.read_csv("earthquakes-1.csv")
```

```python
# Filter data for magType 'mb'
earthquake_mb = earthquake_data[earthquake_data['magType'] == 'mb']

# Calculate correlation coefficients
corr = earthquake_mb[['mag', 'tsunami']].corr()

# Create heatmap
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Coefficients between Earthquake Magnitude and Tsunami (magType: mb)')
plt.show()
```

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 6))

# Box plot for Volume Traded
sns.boxplot(y=fb_data['volume'], ax=axes[0])
axes[0].set_title('Box Plot of Facebook Volume Traded')

# Box plot for Closing Prices
sns.boxplot(y=fb_data['close'], ax=axes[1])
axes[1].set_title('Box Plot of Facebook Closing Prices')

plt.show()
```

```python
fig, ax = plt.subplots()

sns.boxplot(y=fb_data['close'], ax=ax)
q1 = fb_data['close'].quantile(0.25)
q3 = fb_data['close'].quantile(0.75)
iqr_val = iqr(fb_data['close'])
lower_bound = q1 - 1.5 * iqr_val
upper_bound = q3 + 1.5 * iqr_val

# Fill area between bounds
ax.fill_betweenx([lower_bound, upper_bound], 0, 1, alpha=0.2, color='grey')

plt.title('Box Plot of Facebook Closing Prices with Tukey Fence')
plt.show()
```

```python
# Convert date to datetime format
fb_data['date'] = pd.to_datetime(fb_data['date'])

fig, ax = plt.subplots()

ax.plot(fb_data['date'], fb_data['close'], label='Closing Price')

# Shade rectangle
ax.axvspan('2018-07-25', '2018-07-31', color='red', alpha=0.3)

plt.title('Facebook Closing Price with Large Decline Shaded')
plt.xlabel('Date')
plt.ylabel('Closing Price')
plt.legend()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load data
fb_data = pd.read_csv("fb_stock_prices_2018.csv")

# Convert dates to datetime objects
fb_data['date'] = pd.to_datetime(fb_data['date'])

# Create the plot
fig, ax = plt.subplots()

ax.plot(fb_data['date'], fb_data['close'], label='Closing Price')

# Annotations
events = [
    ('Disappointing user growth', '2018-07-25', 170),
    ('Cambridge Analytica', '2018-03-19', 175),
    ('FTC investigation', '2018-03-20', 160)
]

for label, date, y_coord in events:
    date = pd.to_datetime(date)
    ax.annotate(label, xy=(date, fb_data.loc[fb_data['date'] == date, 'close'].iloc[0]),
                xytext=(date, y_coord), arrowprops=dict(facecolor='black', arrowstyle='->'), fontsize=10)

plt.title('Facebook Closing Price with Annotations')
plt.xlabel('Date')
plt.ylabel('Closing Price')
plt.legend()
plt.show()
```
